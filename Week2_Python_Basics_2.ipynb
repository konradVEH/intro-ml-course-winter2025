<span style="font-size: 15px;font-family: Gotham, sans-serif;text-align: left;font-weight: 580;color: #6a4d85;">Handling Categorical Features–</span>In the context of Data Science, we frequently encounter two main types of features: **Numerical** and **Categorical**.

* **Numerical Features** are those that can take on a range of numeric values. For example, in our dataset, `Age` and `Counseling_Hours` are numerical features as they can be expressed as real numbers.

* **Categorical Features**, on the other hand, can only assume a limited set of distinct values. In our dataset, `Sex` is an example of a categorical feature, representing either `Male` or `Female`.

However, most machine learning algorithms require numerical input and output variables. Thus, we need to convert categorical data into a format that can be utilized by machine learning algorithms to enhance prediction accuracy.

* **Label Encoding:** One approach involves converting each value in a column into a number. For example, `Male` could be represented as `1`, and `Female` as `0`. However, this method has a drawback: the numeric values may erroneously imply a hierarchical or ordinal relationship.

* **One-Hot Encoding:** This method is often more suitable for categorical data without a natural order. With one-hot encoding, we create a new binary feature for each category in a feature, indicating the presence (1) or absence (0) of that category. For instance, if our dataset has a `Color` feature with values 'Red', 'Blue', and 'Green', one-hot encoding would produce three new columns: `Is_Red`, `Is_Blue`, and `Is_Green`, each with binary values.

The choice of encoding technique largely depends on the nature of the categorical data at hand. If there is an inherent order (ordinal data), label encoding might be appropriate. Otherwise, especially when avoiding unintended order assumptions, one-hot encoding is often preferred. However, caution should be exercised with one-hot encoding, particularly for features with numerous categories, as *it can significantly increase the dataset's dimensionality*.

**Applying Label Encoding to the 'Sex' Feature:** In our dataset, since `Sex` is binary (Male/Female), label encoding can be safely employed. The `map()` function is typically used to map values of a **Series** from one domain to another. It is commonly used for mapping categorical strings to numerical values. We can also use the `replace()` function. This function is more general and versatile than `map()`. It is used to replace specified values with others, and it can operate on both `DataFrames` and `Series`. While map() is strict and will map all values even if some were not specified (turning them into NaN), replace() only modifies the values specified, leaving others unchanged. While `map()` operates rigorously and assigns a mapping to all values, including those not explicitly specified (converting them to `NaN`), `replace()` selectively alters only the specified values, leaving the others unaffected.

````python
# Using the map() function to transform the 'Sex' column
df['Sex'] = df['Sex'].map({'Male': 0, 'Female': 1})

# Display the first few rows to see the transformation
print(df.head())
````

````python
#Or you could run this instead of the above snippet.
# Using the replace() function to transform the 'Sex' column
df['Sex'] = df['Sex'].replace({'Male': 0, 'Female': 1})

# Display the first few rows to see the transformation
print(df.head())
````

For now, we will go back to the "Male" and "Female" labels.

````python
df['Sex'] = df['Sex'].map({0: 'Male', 1: 'Female'})
# Display the first few rows to see the transformation
print(df.head())
````

After understanding the distribution of our data with histograms, it is often beneficial to explore the specifics of data spread, central tendency, and potential outliers. One of the most effective visual tools for this purpose is the **box plot** (or whisker plot).

A box plot gives a five-number summary of a dataset:
- The minimum (smallest observation)
- The first quartile (Q1 or the 25th percentile)
- The median (Q2 or the 50th percentile)
- The third quartile (Q3 or the 75th percentile)
- The maximum (largest observation)

Any data point that falls below $Q1 - 1.5 \times \text{IQR}$ or above $Q3 + 1.5 \times \text{IQR}$ is considered an outlier, where IQR (Inter-Quartile Range) is $Q3 - Q1$.

Using `Plotly`, we have the capability not only to create interactive visualizations of these summaries but also to hover over the plot and obtain exact values, which proves particularly valuable for thorough data exploration.

````python
# Create an interactive box plot for the 'Age' feature
fig_box = px.box(df, x='Sex', y='Age', title='Box Plot of Age Grouped by Sex', color='Sex', color_discrete_sequence=['#0d7b94', '#82153d'])

# Customizing the layout and theme
fig_box.update_layout(template="plotly_white")

# Show the plot
fig_box.show('colab')
````

1. **Plot the Box Plot**:
   - `px.box()`: This function creates a box plot.
     - `df`: Specifies our dataset.
     - `x='Sex'`: Determines the data column we want on the x-axis. Grouping by 'Sex' means we'll have separate box plots for each gender.
     - `y='Age'`: Determines the data column for the y-axis to visualize its distribution.
     - `title='Box Plot of Age Grouped by Sex'`: Sets the title.
     - `color='Sex'`: Colors the box plots based on the 'Sex' column, providing separate colors for each gender.
     - `color_discrete_sequence=['#0d7b94', '#82153d']`: Defines the colors for the unique values in the 'Sex' column, in this case, male and female. You can replace 'blue' and 'pink' with any other colors of your choice.

2. **Customize the Theme**:
   - `fig_box.update_layout(template="plotly_white")`: This function updates the layout of the plot. The `template="plotly_white"` argument provides a clean white background. As before, you can adjust this theme as per your preference.

3. **Display the Visualization**:

<font color='#2a9d8f' size=5>Indexing with '.iloc' and '.loc' in Pandas</font>

When it comes to working with data in pandas, one of the fundamental tasks is selecting or accessing specific rows, columns, or cells within our DataFrame. To achieve this with flexibility and efficiency, pandas offers two primary methods: `.iloc` and `.loc`. Let's explore these techniques using our mental health dataset.

**1. Introduction to `.iloc`**: `.iloc` predominantly relies on integer-based indexing, meaning it uses integer values to pinpoint data within the DataFrame. Think of it as choosing data based on its position within rows and columns.


````python
# Select the first row
first_row = df.iloc[0]

# Select the first five rows
first_five_rows = df.iloc[0:5]

# Select the first five rows and first two columns
subset = df.iloc[0:5, 0:2]

print(f"This is the first row:\n{first_row}")

print(f"This is the first five rows:\n{first_five_rows}")

print(f"This is the first five rows and first two columns:\n{subset}")
````

- `df.iloc[0]`: Picks the initial row of the DataFrame.
- `df.iloc[0:5]`: Retrieves rows starting from position 0 up to (but not including) position 5.
- `df.iloc[0:5, 0:2]`: The first part `0:5` pertains to row selection, while the second part `0:2` pertains to column selection. This results in a subset of the first five rows and the first two columns.

**2. Introduction to `.loc`**: In contrast to `.iloc` which employs positions, `.loc` operates based on labels, utilizing the names or labels of rows and columns for data selection. When the DataFrame employs a custom index, such as names or dates, `.loc` proves especially useful.

````python
# Setting 'Sex' as our index for demonstration
df_indexed = df.set_index('Sex')
#print(f"This is how the indexed dataset looks like:\n{df_indexed}")

# Select rows for 'Male'
males_data = df_indexed.loc['Male']
print(f"This is the male dataset:\n{males_data}")

# # Select data for 'Male' and specific columns 'Age' and 'Stress Level'
subset_males = df_indexed.loc["Male", ['Age', 'Stress_Score']]
print(f"This is the male dataset, with two columns:\n{subset_males.head()}")
````

- `df.set_index('Sex')`: We're setting the 'Sex' column as our index to demonstrate `.loc` effectively.
- `df_indexed.loc['Male']`: Selects all rows with the index label 'Male'.
- `df_indexed.loc['Male', ['Age', 'Stress Level']]`: Selects rows with the index label 'Male' and only the 'Age' and 'Stress Level' columns.

**3. Tips and Considerations**:
- Always remember, `.iloc` is for **integer-based** location indexing, and `.loc` is for **label-based** indexing.
- If you are unsure about the datatype of the index, use `df.index` to check.
- While it is common to have an integer index, do not confuse this with the positions in `.iloc`. Even if you have a DataFrame with an integer index starting from 100, `df.iloc[0]` will still give you the first row, but `df.loc[0]` will result in an error.
